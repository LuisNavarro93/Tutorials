### Eigendecomposition
$$\textbf{Av}=\lambda\textbf{v}$$

```python
I = np.zeros((3,3),dtype=int)
I[0,0]=1 
I[1,1]=1  
I[2,2]=1
print ("I=","\n",I,"\n")
B = np.diag((1,2,3))
print("B=","\n",B,"\n")
C = np.diag(np.arange(1,4))
print("C=","\n",C)
```

```python
#Sean los Eigenvalores y Eigenvectores, EVal y Evec respectivamente 
```

```python
EVal,EVec=np.linalg.eig(B)
print("Eigenvalores de B:",EVal)
print("Eigenvector de B:","\n",EVec)
#Los eigenvectores se almacenan en matriz como vectores columna 
```

```python
EValC,EVecC=np.linalg.eig(C)
print("Eigenvalores de C:",EValC)
print("Eigenvector de C:","\n",EVecC)
#Los eigenvectores se almacenan en matriz como vectores columna 
```

```python
EVecC
```

```python
print('Eigenvalues : {}'.format(EValC[1]))#Eigenvalor
print('Eigenvector : {}'.format(EVecC[0]))#Vector columna
```

$$\textbf{A}=\textbf{V}diag(\lambda)\textbf{V}^{-1}$$
Sea $\lambda$ los eigenvalores, y sea $\textbf{V}$ una matriz con un eigenvector por columna 

```python
np.diag(EValC) #crear matriz diagonal con valores especificos
```

```python
np.linalg.inv(EVecC) #obtener matriz inversa
```

```python
CMatrix = np.matmul(np.diag(EValC),np.linalg.inv(EVecC))
output = np.matmul(EVecC,CMatrix).astype(np.int)
print(output)
```
