```python
x1=np.linspace(0,10)
print("x =",x,"\n","dimension:",np.shape(x1))
y1=np.sin(x1)
y2=np.cos(x1)
print("y1 =",y1,"\n","dimension:",np.shape(y1))
plt.plot(x1,y1,label="Sin(x)")
plt.plot(x1,y2,label="(Cos(x)")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Trigonometricas")
```

```python
x1=np.linspace(0,10)
print("x =",x,"\n","dimension:",np.shape(x1))
y1=np.sin(x1)
y2=np.cos(x1)
print("y1 =",y1,"\n","dimension:",np.shape(y1))
plt.scatter(x1,y1,label="Sin(x)",c="blue")
plt.scatter(x1,y2,label="(Cos(x)",c="green",marker="*")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Trigonometricas")
```
