```python
x=np.linspace(0,10,100)
y=1/2*np.tan(x)
y1=3*np.tan(x)
fig=plt.figure(figsize=(10,6))
plt.style.use('default')
plt.plot(x,y,label="1/2*tan(x)")
plt.plot(x1,y1,label="3*tan(x)")
plt.legend()

plt.axvline(2)
plt.axvline(5)

plt.fill_between(x,y1,y,y<y1,interpolate=True,facecolor="r")
plt.fill_between(x,y1,y,y>y1,interpolate=True,facecolor="g")

plt.text(2,-100,"3*tan(x)",bbox=dict(boxstyle="square,pad=1",fc="white"))
```
