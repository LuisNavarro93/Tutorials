```python
x=[3,4,6,8,9]
y=[9,6,4,3,2]
plt.plot(x,y,'--r')
plt.xlabel('random x')
plt.ylabel('randome y')
plt.title('Grafica2')
```

```python
x=[3,4,6,8,9]
y=[9,6,4,3,2]
x1=[13,14,16,18,19]
y1=[19,16,14,13,12]
plt.plot(x,y,'--r',label='1erGrafica')
plt.xlabel('random x')
plt.ylabel('randome y')
plt.title('Grafica3')
plt.plot(x,y1,'-*k',label='2daGrafica')
plt.plot(x1,y,'-*c',label='3erGrafica')
#plt.legend??
plt.legend(loc=0)
```

```python
x=np.array((1,2,3,4))
y=np.array((10,20,30,40))
plt.plot(x,y,label='1era')

x1=np.array([10,20,30,40])
y1=np.array([1,2,3,4])
plt.plot(x1,y1,label='2da')

plt.legend(loc=0)
```
